# Intregrar Bases de Datos

```python
# Cargar librerias
import pandas as pd
import numpy as np
import os
```

```python
# Cargar bases de datos
ind_corrupcion = pd.read_csv('Datos/indicadores-corrupcion.csv', sep = ";")
var_secop = pd.read_csv('Datos/variables-secop.csv.gz', sep=';',
                        dtype={'dep_codigo_divipola':'int64', 'municipio_codigo_divipola':'int64'})
ind_edu_mun = pd.read_csv("Datos/educacionMun.csv", dtype={'Código Entidad':'int64'})
ind_edu_dep = pd.read_csv("Datos/educacionDep.csv", dtype={'Código Entidad':'int64'})
#departamentos = pd.read_csv("Datos/cod_divipola_departamento.csv", sep=";")
#municipios = pd.read_csv("Datos/cod_divipola_municipio.csv", sep=";")
#departamentos.head(), municipios.head()
```

```python
# Integrar variables del secop con indicadores de riesgo de corrupción
bases_integradas = pd.merge(left=var_secop, right=ind_corrupcion, on='nombre_entidad', how='inner')
```

```python
# Integrar los indicadores de educación municipal
bases_integradas = bases_integradas.merge(right=ind_edu_mun, left_on='municipio_codigo_divipola', 
                                          right_on='Código Entidad', how='inner', )
```

```python
# Integrar los indicadores de educación departamental
bases_integradas = bases_integradas.merge(right=ind_edu_dep, left_on='dep_codigo_divipola', 
                                          right_on='Código Entidad', how='inner', suffixes=['_mun','_dep'],)
```

```python
eliminar = ['Código Entidad_mun','Entidad_mun', 'Código Entidad_dep', 'Entidad_dep']
bases_integradas.drop(columns=eliminar, inplace=True)
```
